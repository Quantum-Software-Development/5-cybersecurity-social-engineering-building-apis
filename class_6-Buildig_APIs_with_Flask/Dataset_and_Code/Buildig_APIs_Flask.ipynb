```python
import pandas as pd
from flask import Flask
from flask import request,Response
import os

dataFrame=pd.read_excel('Farmacia.xlsx')
dataFrame

app = Flask(__name__)

## TESTAR COM: http://127.0.0.1:5000/farmacia
@app.route("/farmacia",methods = ['GET'])
def get_farmacia():
    return Response(dataFrame.to_json(orient="records"), mimetype='application/json')

## TESTAR COM: http://127.0.0.1:5000/remedio?cod=C104
@app.route("/remedio",methods = ['GET'])
def get_remedio():
    cod = request.args.get('cod')
    prod=dataFrame.loc[dataFrame['Codigo']==cod]
    return Response(prod.to_json(orient="records"), mimetype='application/json')


@app.route("/totalRemedio",methods = ['GET'])
def get_totalRemedio():
    cod = request.args.get('cod')
    prod=dataFrame.loc[dataFrame['Codigo']==cod]
    prod=prod.groupby(['Codigo'])['Pedido'].count().reset_index()
    return Response(prod.to_json(orient="records"), mimetype='application/json')

@app.route("/totalPedido",methods = ['GET'])
def get_totalPedido():
    pedido = request.args.get('pedido')
    prod=dataFrame.loc[dataFrame['Pedido']==int(pedido)]
    prod=prod.groupby(['Pedido'])['Qtde'].sum().reset_index()
    return Response(prod.to_json(orient="records"), mimetype='application/json')

if __name__=="__main__":
    app.run(host=os.getenv('IP', '0.0.0.0'), 
            port=int(os.getenv('PORT', 5000)))
```
